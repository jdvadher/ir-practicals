```python
#importing libraries and global variables
import pandas as pd
import numpy as np
from glob import glob
import re
from os import listdir
from collections import Counter
from tqdm import tqdm
```

```python
directories = []
for i in glob("masc_500k_texts/masc_500k_texts/*/"):
  for j in glob(i+'*/'):
    directories.append(j)
```

```python
directories
```

```python
data = []
for i in directories:
  for j in listdir(i):
    with open(i+j, "r") as f:
      try:
        d = {'fileName': j, 
             'text': f.read()}
        data.append(d)
      except:
        continue
```

```python
len(data)
```

```python
#we need just the words. so remove all the special characters
regex = r"@\S+|https?:\S+|http?:\S|[^A-Za-z0-9 '.]+" #anything that is not a number or alphabet
words = []
for i in data:
  i['text'] = re.sub(regex,' ',i['text'].lower())
  for j in i['text'].split():
    words.append(j.lower())

words = list(set(words))
words.append('<S>')
words.append('<E>')
```

```python
len(words)
```

```python
bigram = pd.DataFrame(data = np.zeros((len(words), len(words)), dtype = 'float16'), columns = words, index=words, dtype = 'float16')
```

```python
bigrams = []
for i in range(len(data)):
  s = data[i]['text'].lower().split()
  s.insert(0,'<S>')
  s.append('<E>')
  for j in range(len(s)-1):
    bigrams.append([s[j],s[j+1]])
#unique_bigrams = []
#[unique_bigrams.append(x) for x in tqdm(bigrams) if x not in unique_bigrams] 
```

```python
for i in tqdm(bigrams):
  bigram[i[1]][i[0]]+=1

bigram
```

```python
bigram = bigram.replace(0.0, 1.0)
bigram.head()
```

```python
bigram["sum"] = bigram.sum(axis=1)
```

```python
for i in bigram.index:
    bigram[i][:] /= bigram['sum'][i]
```
